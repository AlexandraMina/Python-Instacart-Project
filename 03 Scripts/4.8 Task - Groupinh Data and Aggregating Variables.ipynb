# Step 4

```python
# creating a loyalty flag for existing customers using the transform() function
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking
ords_prods_merge.head()
```

```python
ords_prods_merge['max_order'].value_counts()
```

```python
# creating a loyalty flag for existing customers using the loc() function with same criteria in the exercise
# If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
# If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
# If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking 
ords_prods_merge.head()
```

```python
ords_prods_merge['loyalty_flag'].value_counts()
```

# Step 5

```python
# Check the basic statistics of the product prices for each loyalty category
# use groupby and agg() 
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Loyal customers pay a average lower price than the other customers (new customer and regular customer). Loyal customers pay on average 3 dolares less than new customers.

# Step 6

```python
 # Create a spending flag for each user based on the average price across all their orders using the following criteria:
 # If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
 # If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”
```

```python
# Create a column for average order
ords_prods_merge['avg_spender'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check the new column 'avg_spender'
ords_prods_merge.head()
```

```python
ords_prods_merge['avg_spender'].value_counts()
```

```python
# Create a flag for low and high spender
ords_prods_merge.loc[ords_prods_merge['avg_spender'] < 10, 'spender_flag'] = 'Low_spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_spender'] >= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check new spender flag
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['spender_flag'].value_counts()
```

# Step 7

```python
# Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
#If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

```python
# Create a column for median in the “days_since_prior_order” column
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check the new column 'median_days_since_prior_order'
ords_prods_merge.head()
```

```python
ords_prods_merge['median_days_since_prior_order'].value_counts()
```

```python
# Create a flag for Non-frequent customer, Regular customer and Frequent customer
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check new spender flag
ords_prods_merge.head()
```

```python
ords_prods_merge['frequency_flag'].value_counts()
```

```python
# Additional step
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```
